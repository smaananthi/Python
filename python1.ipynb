## iloc vs String

```python
a="nila"
print(a[3]) #identifing the element base on index
print(a[2:]) #grab the remaing elements except upto the index
print(a[:2]) #grab the elements upto the index
print(a[-1]) #grab the last element
print(a[:-1]) #grab the element except last element
print(a[::-2]) #grab everything with 2 steps
print(a[::-1])  #print string backwards
```

```python
b="157"
print(b[:1]) #first value
print(b[:2]) #first two values
print(b[1:]) #except first value
print(b[2:]) #except first two
print(b[:-1]) #except last one
print(b[:-2]) #except last two
print(b[-1:]) #last element
print(b[-2:]) # last two 
print(b[::-1])
print(b[::-2])
```

## Set & Tuple & list & Dictonary

#### Set
1. unordered collection of data
2. curly braces{}
3. mutable | no duplicates
4. use add command
5. pop- remove the random element
6. cannot be sorted
7. cannot get index
8. cannot be reversed

```python
# create a set of string type
set={'a','e','i','o','u'}
print('set=',set)
```

```python
set.add('b'+'z')
set
```

```python
set.add('p')
set
```

```python
# create a set of integer type
student_id = {112, 114, 116, 118, 115}
print('Student ID:', student_id)
```

```python
# create a set of mixed data types
mixed_set = {'hi', 1.2, -2, 'to'}
print('Set of mixed data types:', mixed_set)
```

```python
# first set
A = {1, 3, 5}

# second set
B = {0, 2, 4}

# perform union operation using |
print('Union using |:', A | B)

# perform union operation using union()
print('Union using union():', A.union(B))
```

```python
# first set
A = {1, 3, 5}

# second set
B = {1, 2, 3}

# perform intersection operation using &
print('Intersection using &:', A & B)

# perform intersection operation using intersection()
print('Intersection using intersection():', A.intersection(B)) 
```

```python
# first set
A = {2, 3, 5}

# second set
B = {1, 2, 6}

# perform difference operation using &
print('Difference using &:', A - B)

# perform difference operation using difference()
print('Difference using difference():', A.difference(B)) 
```

```python
# first set
A = {2, 3, 5}

# second set
B = {1, 2, 6}

# perform difference operation using &
print('using ^:', A ^ B)

# using symmetric_difference()
print('using symmetric_difference():', A.symmetric_difference(B)) 
```

```python
# first set
A = {1, 3, 5}

# second set
B = {3, 5, 1}

# perform difference operation using &
if A == B:
    print('Set A and Set B are equal')
else:
    print('Set A and Set B are not equal')
```

```python
print(len(t))
```

#### list
1. collection of ordered data
2. square braces[]
3. mutable - we can remove or add elements to it
4. append-add element
5. pop- remove the element
6. sort- sorting elements
7. index-searching given element
8. reverse- reverse the element

```python
list = ["apple", "banana", "cherry"]
print(list)
```

```python
#allow duplicate
list = ["apple", "banana", "cherry","apple"]
print(list)
```

```python
print(len(list))
```

#### Dictonary
1. unordered collection which stores data in KEY,VALUE Basis
2. curly braces{}
3. mutable | no duplicates
4. update- add value 
5. pop-remove the specified element
6. sorted - sorting keys DEFAULT
7. get - return value with element key
8. cannot be reversed

```python
dict = {
  "brand": "hero",
  "model": "345",
  "year": 2000
}
dict
```

```python
print(dict['brand'])
```

```python
#add duplicate
dict = {
  "brand": "hero",
  "model": "4582",
  "year": 2000,
  "year": 2020
}
print(dict)
```

```python
#changing in dic
a={"Name":"Elon", "Age":50,"company":["spaceX","Tesla"]}
print(a["Name"]) #printing value based on its key
a["Age"]=51
a
```
